```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import math
import warnings
warnings.filterwarnings('ignore')
```

```
pd.options.display.float_format = '{:20,.2f}'.format
data = pd.read_csv('dmi_cibil_report.csv')
```

```
data.set_index('lead_id', inplace=True)
original_data = data.copy()
data.head()
```

```
data = original_data.copy()
old_columns = list(data)
for i in range(62):
  data['total_'+str(i)] = 0
  data['max_'+str(i)] = 0
  data['count_'+str(i)] = 0
  data['etotal_'+str(i)] = 0
  data['emax_'+str(i)] = 0
  data['ecount_'+str(i)] = 0
```

```
print(data.shape)
print(list(data))
```

```
data.head()
```

```
for index,row in data.iterrows():
  for i,j in enumerate(old_columns):
    am = j.replace('type','amountAmount')
    em = j.replace('type','emiAmount')
    if 'type' in j and not math.isnan(row[am]):
      a = str(int(row[old_columns[i]]))
      total_column = "total_"+a
      max_column = "max_"+a
      count_column = "count_"+a
      data.at[index, total_column] += row[am]
      data.at[index, max_column] = max(data.at[index, max_column], row[am])
      data.at[index, count_column] += 1
      if j.replace('type','emiAmount') in old_columns and not math.isnan(row[em]):
        data.at[index, "etotal_"+a] += row[em]
        data.at[index, "emax_"+a] = max(data.at[index, "emax_"+a], row[em])
        data.at[index, "ecount_"+a] += 1  

data.head()
```

```
data.to_csv('new_file_emi.csv')
```
